```python
import os
import json 

# Set your directory path
directory_path = os.path.join(os.getcwd(), 'Features/FeaturesNoECG')
# directory_path = os.path.join(base_path)
data = {}
names = []
# Check if the directory exists
if os.path.exists(directory_path):
    # List all files in the directory
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        name = filename.split('.')[0]
        # Check if it is a file
        if os.path.isfile(file_path):
            print(f"Reading file: {filename}")
            try:
                with open(file_path, 'r') as file:
                    data[name] = json.load(file)
                    names.append(name)
            except Exception as e:
                print(f"Error reading file {filename}: {e}")
else:
    print("The directory does not exist.")
```

```python
import ast
from collections import defaultdict

combined_list = []
for nam in names:
    primary_list = ast.literal_eval(data[nam]['Video Condition'])
    combined_list.append(primary_list)

print(combined_list)
```

```python


defaultdicts = []

for data_sample in combined_list:
    indices_of_duplicates = defaultdict(list)

    for index, value in enumerate(data_sample):
        indices_of_duplicates[value].append(index)

    defaultdicts.append(indices_of_duplicates)
    
common_keys = set(defaultdicts[0].keys())  # Start with keys from the first defaultdict
for ddict in defaultdicts[1:]:  # Iterate over the rest of the defaultdicts
    common_keys.intersection_update(ddict.keys())
    
    
combined_indices = defaultdict(list)

for key in common_keys:
    for ddict in defaultdicts:
        combined_indices[key].extend(ddict[key])

print(combined_indices)
```

```python
def segment_sequences(df, sequence_length):
    """
    Segments the DataFrame into smaller sequences of a given length,
    maintaining the label for each sequence.
    
    Parameters:
    - df: DataFrame containing features and labels.
    - sequence_length: Length of each segmented sequence.
    
    Returns:
    - Segmented feature set and corresponding labels as arrays.
    """
    segmented_X, segmented_y = [], []
    
    # Iterate over each unique label
    for label in df['label'].unique():
        subset = df[df['label'] == label]  # Filter data for the current label
        X = subset.drop('label', axis=1).values  # Features
        y = subset['label'].values[0]  # Label (assuming all rows in subset have the same label)
        
        # Break the subset into segments of the desired length
        for i in range(0, len(X) - sequence_length + 1, sequence_length):
            segment_X = X[i:i+sequence_length]
            segmented_X.append(segment_X)
            segmented_y.append(y)  # Each segment has the same label
    
    return np.array(segmented_X), np.array(segmented_y)

```
