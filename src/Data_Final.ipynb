```python
import os
import json 

# Set your directory path
directory_path = os.path.join(os.getcwd(), 'Features')
# directory_path = os.path.join(base_path)
data = {}
names = []
# Check if the directory exists
if os.path.exists(directory_path):
    # List all files in the directory
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        name = filename.split('.')[0]
        # Check if it is a file
        if os.path.isfile(file_path):
            print(f"Reading file: {filename}")
            try:
                with open(file_path, 'r') as file:
                    data[name] = json.load(file)
                    names.append(name)
            except Exception as e:
                print(f"Error reading file {filename}: {e}")
else:
    print("The directory does not exist.")
```

```python
import ast
from collections import defaultdict

combined_list = []
for nam in names:
    primary_list = ast.literal_eval(data[nam]['Video Condition'])
    combined_list.append(primary_list)

print(combined_list)
```

```python


defaultdicts = []

for data_sample in combined_list:
    indices_of_duplicates = defaultdict(list)

    for index, value in enumerate(data_sample):
        indices_of_duplicates[value].append(index)

    defaultdicts.append(indices_of_duplicates)
    
common_keys = set(defaultdicts[0].keys())  # Start with keys from the first defaultdict
for ddict in defaultdicts[1:]:  # Iterate over the rest of the defaultdicts
    common_keys.intersection_update(ddict.keys())
    
    
combined_indices = defaultdict(list)

for key in common_keys:
    for ddict in defaultdicts:
        combined_indices[key].extend(ddict[key])

print(combined_indices)
```

```python
ECG_feature_names = data['Aaron Fainman_Features']['ECG Features'][0].keys()
ECG_feature_names = list(ECG_feature_names)[1:]

RSP_feature_names = data['Aaron Fainman_Features']['RSP Features'][0].keys()
RSP_feature_names = list(RSP_feature_names)[1:]

PPG_feature_names = data['Aaron Fainman_Features']['PPG Red Features'][0].keys()
PPG_feature_names = list(PPG_feature_names)[1:]

EDA_feature_names = data['Aaron Fainman_Features']['EDA Features'][0].keys()
EDA_feature_names = list(EDA_feature_names)[1:]

all_feature_names = ECG_feature_names + RSP_feature_names + PPG_feature_names + EDA_feature_names

data_columns = all_feature_names + ['label']

feature_type_list = {
    'ECG Features':  ECG_feature_names,
    'RSP Features': RSP_feature_names,
    'PPG Red Features': PPG_feature_names,
    'EDA Features': EDA_feature_names
}
```

```python
# Find indices of duplicates
import ast
from collections import defaultdict
import pandas as pd
# Initialize an empty DataFrame to store categorized data
# Construct empty data frame with columns
import numpy as np

df = pd.DataFrame(columns=data_columns)

Features = ['ECG Features', 'RSP Features', 'PPG Red Features', 'EDA Features']

video_to_quandrant = {}

# Mapping from labels to columns
label_to_column = combined_indices

for iname, nam in enumerate(names):
    for feat in Features:
        for i in range(15):
            if feat in data[nam]:
                video_name = data[nam][feat][i]['Label'].split('.')[0]
                quadrant = combined_list[iname][i]
                video_to_quandrant[video_name] = quadrant

idx = 0
for iname, nam in enumerate(names):
    for i in range(15):
        for feat in Features:
            if feat in data[nam]:
                feature_data = data[nam][feat][i]
                for key in feature_data.keys():
                    if key != 'Label':
                        df.loc[idx, key] = feature_data[key]
            else:
                for key in feature_type_list[feat]:
                    df.loc[idx, key] = np.nan
        
        video_name = data[nam]['PPG Red Features'][i]['Label'].split('.')[0]
        df.loc[idx, 'label'] = video_to_quandrant[video_name]        
        idx += 1   
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer
import pandas as pd

# Define components, imputer, and scaler
n_components = 50
imputer = SimpleImputer(strategy='median')
robust_scaler = RobustScaler()
pca = PCA(n_components=n_components)

# Make a pipeline
pipeline = make_pipeline(imputer, robust_scaler, pca)

# Separate features from labels if not already done
features = df.drop(columns=['label'])

# Fit the pipeline and transform the data to n_components dimensions
df_scaled_transformed = pipeline.fit_transform(features)

# Convert the PCA-transformed data to a DataFrame
pca_df = pd.DataFrame(data=df_scaled_transformed, columns=[f'PC{i+1}' for i in range(n_components)])

# Add the labels back to the DataFrame
pca_df['label'] = labels.values  # Make sure labels align with the rows in pca_df

# Save the DataFrame as JSON
pca_df.to_json('pca_data_with_labels.json', orient='records')


import matplotlib.pyplot as plt
import numpy as np

# Assuming 'labels' contains categorical data for coloring
# Recreate the color mapping if needed
unique_labels = np.unique(labels)
label_color_map = {label: idx for idx, label in enumerate(unique_labels)}
label_colors = [label_color_map[label] for label in labels]

# Creating a 3D scatter plot with colors for the first 3 components
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'], c=label_colors, cmap='viridis')

# Create a legend
# This creates a legend with a color for each label
handles, _ = scatter.legend_elements(prop='colors')
legend_labels = [f'Label {l}' for l in unique_labels]  # Adjust based on your label names
ax.legend(handles, legend_labels, title="Labels")

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Plot with Labels')

plt.show()
```

```python
from sklearn.manifold import TSNE

# Create an imputer object with a median filling strategy
imputer = SimpleImputer(strategy='median')

# Impute missing values in 'df'
df_imputed = imputer.fit_transform(df)

# Create a robust scaler object
robust_scaler = RobustScaler()

# Scale the imputed data
df_scaled = robust_scaler.fit_transform(df_imputed)

tsne = TSNE(n_components=2, random_state=42)  # n_components=2 for 2D visualization

# Apply t-SNE to the scaled data
df_tsne = tsne.fit_transform(df_scaled)

# Creating a color map based on labels
unique_labels = np.unique(labels)
label_color_map = {label: idx for idx, label in enumerate(unique_labels)}
label_colors = np.array([label_color_map[label] for label in labels])

# Plotting the t-SNE results
plt.figure(figsize=(10, 8))
plt.scatter(df_tsne[:, 0], df_tsne[:, 1], c=label_colors, cmap='viridis', alpha=0.7)

# Creating a legend (optional)
# This assumes 'labels' are numeric or have been converted to numeric
handles = [plt.Line2D([0], [0], marker='o', color='w', label=unique_labels[idx], 
                      markerfacecolor=plt.cm.viridis(idx / len(unique_labels)), markersize=10) for idx in range(len(unique_labels))]
plt.legend(handles=handles, title='Labels')

plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.title('t-SNE Visualization')
plt.show()
```
